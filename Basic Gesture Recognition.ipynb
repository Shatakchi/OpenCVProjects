```python
import numpy as np
import cv2 as cv
```

```python
img= cv.imread('UHand.jpg',0)
ori=img
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure()
plt.axis("off")
plt.imshow(img)
plt.show()
```

```python
ret, thre = cv.threshold(img, 170, 255, cv.THRESH_BINARY)
```

```python
plt.figure()
plt.axis("off")
plt.imshow(thre)
plt.show()
```

```python
contours,_ = cv.findContours(thre.copy(),cv.RETR_TREE,cv.CHAIN_APPROX_SIMPLE)
final2= cv.drawContours(img, contours, -1, (0,255,0), 3)
```

```python
plt.figure()
plt.axis("off")
plt.imshow(final2)
plt.show()
```

```python
areas = [cv.contourArea(c) for c in contours]
max_index = np.argmax(areas)
areas.pop(max_index)
cnt=contours[5] #Removing the largest(border) and then using a random one
```

```python
img= cv.imread('UHand.jpg',0)
```

```python
rect = cv.minAreaRect(contours[5])
box = cv.boxPoints(rect)
box = np.int0(box)
final4=cv.drawContours(img,[box],0,(0,0,255),2)
```

```python
plt.figure()
plt.axis("off")
plt.imshow(final4)
plt.show()
```
