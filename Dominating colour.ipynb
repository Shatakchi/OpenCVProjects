```python
import numpy as np
import cv2 as cv
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img=cv.imread("Fruits.jpg")
plt.imshow(img)
```

```python
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img = img.reshape((img.shape[0] * img.shape[1],3))
```

```python
kmeans = KMeans(n_clusters = 5)
```

```python
kmeans.fit(img)
```

```python
kmeans.cluster_centers_
```

```python
labels = kmeans.fit_predict(img)
```

```python
from collections import Counter
```

```python
kmeans.labels_
```

```python
count = Counter(labels) 
```

```python
count
```

```python
h=count.most_common(1)[0][0]
h
```

```python
dc = kmeans.cluster_centers_[h]
```

```python
dc
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
bar = np.zeros((3,3,3), dtype="uint8")
```

```python
cv.rectangle(bar,(0,0),(350,350),dc, -1)
```

```python
print("Most dominating color is:")
plt.figure()
plt.axis("off")
plt.imshow(bar)
plt.show()
```
